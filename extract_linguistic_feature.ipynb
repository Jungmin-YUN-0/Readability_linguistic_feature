```python
import pandas as pd

train = pd.read_csv(r'./onestopeng/train.csv', encoding='UTF-8')
train.head()
test = pd.read_csv(r'./onestopeng/test.csv', encoding='UTF-8')
```

```python
lfs = []

def LF(text):
  LingFeat = extractor.pass_text(text)
  LingFeat.preprocess()

  EnDF = LingFeat.EnDF_() # Entity Density Features
  EnGF = LingFeat.EnGF_() # Entity Grid Features

  # Syntactic (Synta) Features
  PhrF = LingFeat.PhrF_() # Noun/Verb/Adj/Adv/... Phrasal Features
  TrSF = LingFeat.TrSF_() # (Parse) Tree Structural Features
  POSF = LingFeat.POSF_() # Noun/Verb/Adj/Adv/... Part-of-Speech Features
  
  # Lexico Semantic (LxSem) Features
  TTRF = LingFeat.TTRF_() # Type Token Ratio Features
  VarF = LingFeat.VarF_() # Noun/Verb/Adj/Adv Variation Features 
  PsyF = LingFeat.PsyF_() # Psycholinguistic Difficulty of Words (AoA Kuperman)
  WoLF = LingFeat.WorF_() # Word Familiarity from Frequency Count (SubtlexUS)

  # Shallow Traditional (ShTra) Features
  ShaF = LingFeat.ShaF_() # Shallow Features (e.g. avg number of tokens)
  TraF = LingFeat.TraF_() # Traditional Formulas
  lf = []
  lf.append(list(EnDF.values()))
  lf.append(list(EnGF.values()))
  lf.append(list(PhrF.values()))
  lf.append(list(TrSF.values()))
  lf.append(list(POSF.values()))
  lf.append(list(TTRF.values()))
  lf.append(list(VarF.values()))
  lf.append(list(PsyF.values()))
  lf.append(list(WoLF.values()))
  lf.append(list(ShaF.values()))
  lf.append(list(TraF.values()))
  lfs.append(lf)
```

```python
lfs = []

for train in train['text']:
  LF(str(train))
```

```python
train_lf = lfs
train_lf = pd.DataFrame(train_lf)
train_lf.to_csv(r"./onestopeng/linguistic_feature/train_lf.csv", index=False)
```

```python
lfs=[]

for test in test['text']:
  LF(str(test))
```

```python
test_lf = lfs
test_lf = pd.DataFrame(test_lf)
test_lf.to_csv(r"./onestopeng/linguistic_feature/train_lf.csv", index=False)
```
